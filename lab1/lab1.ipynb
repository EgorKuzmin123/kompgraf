```python
from PIL import Image
from math import sqrt, sin, cos, radians
```

Реализовать алгоритм Брезенхема рисования отрезка и окружности, алгоритма цифрового дифференциального анализатора и параметрического алгоритма рисования окружности.

Алгоритм Брезенхема для отрезка

```python
def Brezenheim_otrezok(img, x0, y0, x1, y1):
    b = int((img.size[0] - 1) / 2)
    pixels = img.load()
    steep = abs(y1 - y0) > abs(x1 - x0)
    if steep:
        x0, y0 = y0, x0
        x1, y1 = y1, x1
    if x0 > x1:
        x0, x1 = x1, x0
        y0, y1 = y1, y0
    dx = x1 - x0
    dy = abs(y1 - y0)
    d = 2 * dy - dx
    y = y0
    step_y = 1 if y1 > y0 else -1
    for x in range(x0, x1 + 1):
        curr_x, curr_y = (y, x) if steep else (x, y)
        pixels[b + curr_x, b - curr_y] = 0
        if d > 0:
            y += step_y
            d -= 2 * dx
        d += 2 * dy
```

```python
img = Image.new("L", (41, 41), color=255)
Brezenheim_otrezok(img,-20, 20, 20, 20)
Brezenheim_otrezok(img,20, 20, -20, -20)
Brezenheim_otrezok(img,-20, -20, 20, -20)
img.show()
img.save("output.jpg")
```

Алгоритм Брезенхема для окружности

```python
def Brezenheim_okruznost(img, x0, y0, r):
    b = 20
    pixels = img.load()
    x = 0
    y = 0
    d = 3 - 2 * r
    while (x + x0) <= (x0 + r / sqrt(2)):
        pixels[b + x0 + x, b - y0 - r + y] = 0
        pixels[b + x0 + x, b - y0 + r - y] = 0
        pixels[b + x0 - x, b - y0 - r + y] = 0
        pixels[b + x0 - x, b - y0 + r - y] = 0
        pixels[b + x0 + r - y, b - y0 - x] = 0
        pixels[b + x0 + r - y, b - y0 + x] = 0
        pixels[b + x0 - r + y, b - y0 - x] = 0
        pixels[b + x0 - r + y, b - y0 + x] = 0
        if d < 0:
            d += 4 * x + 6
        else:
            d += 4 * (x - r + y) + 10
            y += 1
        x += 1
```

```python
img = Image.new("L", (41, 41), color=255)
Brezenheim_okruznost(img, 5, -7, 12)
img.show()
img.save("output1.jpg")
```

Цифровой дифференциальный анализатор (ЦДА/DDA)

```python
def DDA(img, x0, y0, x1, y1):
    b = 20
    pixels = img.load()
    delta_x = x1 - x0
    delta_y = y1 - y0
    l = max(abs(delta_x), abs(delta_y))
    dx = delta_x / l
    dy = delta_y / l
    for i in range(l + 1):
        pixels[b + round(x0), b - round(y0)] = 0
        x0 += dx
        y0 += dy
```

```python
img = Image.new("L", (41, 41), color=255)
DDA(img, 0, 0, -15, 7)
img.show()
img.save("output2.jpg")
```

Параметрический алгоритм рисования окружности

```python
def Parametric_ocruznost(img, x0, y0, r):
    b = 20
    pixels = img.load()
    for i in range(361):
        x = x0 + round(r * cos(radians(i)))
        y = y0 + round(r * sin(radians(i)))
        pixels[b + x, b - y] = 0
```

```python
img = Image.new("L", (41, 41), color=255)
Parametric_ocruznost(img, 0, 0, 20)
img.show()
img.save("output3.jpg")
```
