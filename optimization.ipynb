```python
import math
import numpy as np
a_param, b_param = 1.0, 1.0
```

Одномерная минимизация.
Задание 1.

```python
def f_1d(x):
    return a_param / math.exp(x) + b_param * x
```

Метод дихотомии

```python
def dichotomy_method(a, b, eps=1e-5):
    iters = 0
    while (b - a) / 2.0 > eps:
        iters += 1
        delta = eps / 2.0
        x1, x2 = (a + b) / 2.0 - delta, (a + b) / 2.0 + delta
        if f_1d(x1) < f_1d(x2):
            b = x2
        else:
            a = x1
        print(f"итерация: {iters} x* = {(a + b) / 2} f(x*) = {f_1d((a + b) / 2)}")
    return (a + b) / 2.0, iters
```

```python
x_d, i_d = dichotomy_method(0.0, 2.0)
```

Метод золотого сечения

```python
def golden_section_method(a, b, eps=1e-5):
    resphi = 2 - (1 + math.sqrt(5)) / 2 # ~0.382
    x1 = a + resphi * (b - a)
    x2 = b - resphi * (b - a)
    f1, f2 = f_1d(x1), f_1d(x2)
    iters = 0

    while (b - a) > eps:
        iters += 1
        if f1 < f2:
            b = x2
            x2, f2 = x1, f1
            x1 = a + resphi * (b - a)
            f1 = f_1d(x1)
        else:
            a = x1
            x1, f1 = x2, f2
            x2 = b - resphi * (b - a)
            f2 = f_1d(x2)
        print(f"итерация: {iters} x* = {(a + b) / 2} f(x*) = {f_1d((a + b) / 2)}")
```

```python
golden_section_method(0.0, 2.0)
```

Многомерная безусловная минимизация Задание 2

```python
def f1(x):
    return 100 * (x[1] - x[0]**2)**2 + (1 - x[0])**2

def grad_f1(x):
    return np.array([
        -400 * x[0] * (x[1] - x[0]**2) - 2 * (1 - x[0]),
        200 * (x[1] - x[0]**2)
    ])

def f2(x):
    return (x[0]**2 + x[1] - 11)**2 + (x[0] + x[1]**2 - 7)**2

def grad_f2(x):
    return np.array([
        4 * x[0] * (x[0]**2 + x[1] - 11) + 2 * (x[0] + x[1]**2 - 7),
        2 * (x[0]**2 + x[1] - 11) + 4 * x[1] * (x[0] + x[1]**2 - 7)
    ])
```

Градиентный метод

```python
def gradient_descent(func, grad, x0, alpha=0.001, tol=1e-5, max_iter=2000):
    print("\n--- Градиентный метод ---")
    x = np.array(x0, dtype=float)

    for k in range(1, max_iter + 1):
        g = grad(x)
        g_norm = np.linalg.norm(g)
        print(f"Итер: {k:4d} | x=[{x[0]:7.4f}, {x[1]:7.4f}] | f(x)={func(x):9.5f} | ||grad||={g_norm:9.5f}")

        if g_norm < tol:
            break

        x = x - alpha * g
```

```python
x_start_1 = [-1.0, 1.0]
gradient_descent(f1, grad_f1, x_start_1, alpha=0.001)
x_start_2 = [0.0, 0.0]
gradient_descent(f2, grad_f2, x_start_2, alpha=0.01)
```

Метод сопряженных градиентов

```python
def conjugate_gradients(func, grad, x0, alpha=0.001, tol=1e-5, max_iter=2000):
    print("\n--- Метод сопряженных градиентов ---")
    x = np.array(x0, dtype=float)
    g = grad(x)
    p = -g
    for k in range(1, max_iter + 1):
        g_norm = np.linalg.norm(g)
        print(f"Итер: {k:4d} | x=[{x[0]:7.4f}, {x[1]:7.4f}] | f(x)={func(x):9.5f} | ||grad||={g_norm:9.5f}")
        if g_norm < tol:
            break
        x = x + alpha * p
        g_new = grad(x)
        beta = max(0.0, np.dot(g_new, g_new - g) / np.dot(g, g))
        p = -g_new + beta * p
        g = g_new
```

```python
conjugate_gradients(f1, grad_f1, x_start_1, alpha=0.001)
conjugate_gradients(f2, grad_f2, x_start_2, alpha=0.01)
```

Квазиньютоновский метод

```python
def bfgs_method(func, grad, x0, alpha=0.001, tol=1e-5, max_iter=2000):
    print("\n--- Квазиньютоновский метод (BFGS) ---")
    x = np.array(x0, dtype=float)
    I = np.eye(2)
    H = I.copy()  # Приближение обратной матрицы Гессе
    g = grad(x)

    for k in range(1, max_iter + 1):
        g_norm = np.linalg.norm(g)
        print(f"Итер: {k:4d} | x=[{x[0]:7.4f}, {x[1]:7.4f}] | f(x)={func(x):9.5f} | ||grad||={g_norm:9.5f}")

        if g_norm < tol:
            break

        p = -np.dot(H, g)
        x_new = x + alpha * p
        g_new = grad(x_new)

        s = x_new - x
        y = g_new - g

        # Обновление матрицы H
        rho_denom = np.dot(y, s)
        if rho_denom > 1e-10:
            rho = 1.0 / rho_denom
            A1 = I - rho * np.outer(s, y)
            A2 = I - rho * np.outer(y, s)
            H = np.dot(A1, np.dot(H, A2)) + rho * np.outer(s, s)

        x = x_new
        g = g_new
```

```python
bfgs_method(f1, grad_f1, x_start_1, alpha=0.001)
bfgs_method(f2, grad_f2, x_start_2, alpha=0.01)
```
